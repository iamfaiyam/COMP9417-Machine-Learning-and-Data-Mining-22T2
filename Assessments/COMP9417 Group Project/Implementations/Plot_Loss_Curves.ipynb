```python
import re
import matplotlib.pyplot as plt
```

## Plot train loss for different models

```python
with open('/content/losses.log') as fp:
    text = fp.read()

lines = text.split('\n')
loss_lines = [x for x in lines if x]
batches = [re.split(':', str)[0] for str in loss_lines]
losses = [re.split(':', str)[1] for str in loss_lines]
loss_floats = [float(i) for i in losses]
```

```python
plt.figure(figsize = (8, 8))
plt.plot(loss_floats)
plt.title('Training loss')
plt.xlabel('batch')
plt.ylabel('loss')
plt.show()
```

```python
with open('/content/loss_adam.log') as fp:
    text = fp.read()

adam_lines = text.split('\n')
adam_loss_lines = [x for x in lines if x]
adam_batches = [re.split(':', str)[0] for str in loss_lines]
adam_losses = [re.split(':', str)[1] for str in adam_loss_lines]
adam_loss_floats = [float(i) for i in adam_losses]
```

```python
plt.figure(figsize = (8, 8))
plt.plot(adam_loss_floats)
plt.title('Training loss - Adam optimiser')
plt.xlabel('batch')
plt.ylabel('loss')
plt.show()
```
